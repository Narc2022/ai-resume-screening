```python
import re

def cleanResume(text):
    text = str(text)  # ensure text string ho
    
    text = re.sub(r'http\S+', ' ', text)      # remove URLs
    text = re.sub(r'RT|cc', ' ', text)        # remove RT and cc
    text = re.sub(r'#\S+', ' ', text)         # remove hashtags
    text = re.sub(r'@\S+', ' ', text)         # remove mentions
    text = re.sub(r'[^A-Za-z\s]', ' ', text)  # remove special characters
    text = re.sub(r'\s+', ' ', text)          # remove extra spaces
    
    return text.lower().strip()
```

```python
cleanResume("my #### $ # #noorsaeed webiste like is this http://heloword and access it @gmail.com")
```

```python
df['Category'] = df['Category'].fillna('').apply(lambda x: cleanResume(x))


# import pandas as pd

# df = pd.read_excel("your_file.xlsx")
```

```python
df ['Category'][0]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df.Category.unique()
```

VACTORIZATION

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Category'])

requiredText = tfidf.transform(df['Category'])
```

                    SPLITTING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredText, df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
clf = OneVsRestClassifier (KNeighborsClassifier())

clf.fit(X_train, y_train)   # Capital X
ypred = clf.predict(X_test)

print(accuracy_score(y_test,ypred))
```
